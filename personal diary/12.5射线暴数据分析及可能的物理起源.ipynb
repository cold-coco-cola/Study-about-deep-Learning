### 光谱的构成成分和他们的物理原理

##### **构成成分分析**：
1. **band类**：（占多数）能段               
2. **准热成分**：时间变窄时光谱也变“窄”     
3. **PL成分**：延伸至高能区  

##### **可能的物理原理**：
**I**.band：非热机制=>需要存在按照幂律分布的相对论性电子。
存在三个瞬时辐射候选模型：

1）同步辐射；2）同步自康普顿散射；3）热光子源康普顿上散射。

他们都指示了射线暴的喷流物质为磁主导：

理论上预期的三个光谱特征（为什么理论上有？）：准热光球辐射、SSC成分（如果MeV成分是同步辐射起源的话），以及高能的电子-正电子对产生截止，都没有观察到。但是在坡印廷通量主导下能够有合理的解释。
>与耗散光球辐射模型矛盾

**II**.似热成分BB：热机制。=>光球层辐射

**III**.PL成分：非热机制，常与BB成分同时出现；在低能区和高能区谱指数相等。有理论模型认为PL成分是同步辐射、SSC、热光子康普顿化的组合。
问题是，非热成分PL怎么与非热成分band区分？
首先PL与BB成分有高相关性，一种解释是PL就是BB的康普顿上散射成分

**可能性讨论**：
组合这三个东西：

目前出现过的组合有I，II&III，I&III（090926A可能）。

剩下的
1. IIII：II常伴随III，如果III来自II，III应该不会单独存在。
2. I+II：如果BB成分来自光球层辐射，应该会常见，需要确认。
3. I+II+III：要限制的参数很多

### 结论：
1. 存在不同情况:080916和090902出现的相反情况
2. 提出有三个基本构成成分：band；BB；PL
3. band最典型普遍
4. BB成分
5. PL能量在接近高能区释放
1. 理论上，还有两个组合会存在

***物理意义分析*：**
1. Band模型与最简单的火球光球-内部激波模型不一致，可能的解释中，引入坡印廷主导喷流是合理的。而认为这个成分就是光球层辐射是不合理的。
2. 光球成分在090902中出现，以及PL成分的出现，暗示喷出物是多种组成的可能性
3. GeV区域延迟的可能性：（1）最开始超出灵敏度，知道之后指数平缓，可探测（2）喷出物不透明，产生光谱截止。反驳了认为Gev区域成分与Mev区域成分起源不同的观点。
4. Gev辐射成分不太可能来自外部激波（gev是MeV的延伸；在明亮GRB中，LAT与GBM波段活动同步），可能与Mev一样具有内部起源。
