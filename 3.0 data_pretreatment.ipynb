```python
import os


os.makedirs(os.path.join('..', 'data'), exist_ok=True) #创建一个data目录
data_file = os.path.join('..', 'data', 'house_tiny.csv') #定义路径（house_tiny.csv）csv means columns separated
with open(data_file, 'w') as f:                          #open the files , f represent its name
    f.write('NumRooms,Alley,Price\n')                    # 列名
    f.write('NA,Pave,127500\n')                          # 每行表示一个数据样本
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')

import pandas as pd

data = pd.read_csv(data_file)
data

```

```python
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean(numeric_only=True))
print(inputs)
```

`PandaFrame.mean()`会对所有值进行取平，加上`numeric_only=True`可以只取数值的平均。

```python
inputs = pd.get_dummies(inputs, dummy_na=True,dtype=int)
print(inputs)
```

同样地，`get_dummies()`默认为bull值，`dtype=int`可改变数据类型为int，改变数据类型为数值后，之后的操作就能正常进行不会报错。

```python
import torch

X = torch.tensor(inputs.to_numpy(dtype=float))
y = torch.tensor(outputs.to_numpy(dtype=float))
X, y
```

```python
nan_numer = data.isnull().sum(axis=0)  #isnull()&isna() 判断是否为nan并返回布尔值，最后总体返回bull列表，再sum求True的数量
nan_max_id = nan_numer.idxmax()    #找到最大值的索引（index）
data = data.drop([nan_max_id], axis=1) #删除该列（）
data

inputs = data.iloc[:, 0:2]
inputs = inputs.fillna(inputs.mean(numeric_only=True))
print(inputs)
```

`isna()`:判断是否缺失 `idxmax()`:最大值的索引 `axis=0`:对象为0维，在`sum()`中表示把0维（方向）压缩（压缩对象），在`data.drop()`中表示操作对象为列 熟悉这些代码的含义
